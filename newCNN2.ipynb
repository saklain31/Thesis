```python
import numpy as np
temp = []
out = []

for i in range(int(a)):
    temp.append(0)
    tempfin=np.asarray(temp)
    out.append(tempfin)
    temp.clear()
    
for i in range(int(c)):
    temp.append(1)
    tempfin=np.asarray(temp)
    out.append(tempfin)
    temp.clear()
    
for i in range(int(e)):
    temp.append(2)
    tempfin=np.asarray(temp)
    out.append(tempfin)
    temp.clear()


labels_test=np.asarray(out)
print(labels_test)
print(labels_test.size)
print(type(labels_test))

total_labels = np.append(labels,labels_test,axis=0)
print(total_labels.size)
total_labels = total_labels[30000:297949]
print(total_labels.size)
```

```python

import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
# For a single-input model with 10 classes (categorical classification):

model = Sequential()
model.add(Dense(32, activation='relu', input_dim=12))
model.add(Dense(48, activation='softmax'))

model.add(Dense(32, activation='softmax'))
model.add(Dense(64, activation='relu'))

model.add(Dense(3, activation='softmax'))

model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

import numpy as np

total_hot_labels = keras.utils.to_categorical(total_labels, num_classes=3)
model.fit(dataset, total_hot_labels , epochs=30, batch_size=32, validation_split=0.2, verbose=1)

#print(trainset.size/12)
#print(labels.size)
#print(testset.size/12)
#print(labels_test.size)

```
